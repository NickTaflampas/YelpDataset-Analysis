## Preproccessing
The data we have is decent, but for the purposes of guessing a function, we sort our values using A as our index.
Therefore, the new pairs are the same, but the values of A are in order.

```python
A = []
B = []
C = []
file = open("data.csv", "r")
file.readline() #Skip first line

for line in file:
    spl = line.split(",")
    A.append(float(spl[0]))
    B.append(float(spl[1]))
    C.append(float(spl[2]))

#Sorting the A values for Plotting
BFunc = {A[i] : B[i] for i in range(len(A))}
CFunc = {A[i] : C[i] for i in range(len(A))}

sortedB = dict(sorted(BFunc.items()))
sortedC = dict(sorted(CFunc.items()))

A = [i for i in sortedB.keys()]
B = [i for i in sortedB.values()]
C = [i for i in sortedC.values()]

```

## Analysis: A First View
After plotting the data, we can immediatly see an Exponential Relationship on both functions. By tweaking the data, we can even estimate a mathematical formula for each function.
- B(A) = 1.7<sup>A</sup>
- C(A) = 1.55<sup>A</sup>

This of course is just an eye-estimate. And we immediatly see that B function can't be approached with our formula.

```python
#It's not difficult to see that both functions are exponential. However, that is just by looking at them.
#Note the cyan function. That's my eye-estimate function according to my above speculation.
plt.figure(figsize=(20, 6))
plt.subplot(231)
plt.title("Function B(A)")
plt.scatter(A, B)
plt.plot(A, [1.7**i for i in A], color="c")

plt.subplot(132)
plt.title("Function C(A)")
plt.scatter(A, C, color="r")
plt.plot(A, [1.55**i for i in A], color="c")

plt.subplot(133)
plt.title("Both Functions")
plt.scatter(A, B)
plt.scatter(A, C, color="r")
```

## Analysis: Log-Log Plot
Plotting our data in a log-log plot, shows an incredible find. 
- C function continuous to follow an exponential manner, thus proving that it is actually exponential, as we expected.
- B function on the other hand, follows a linear model! This means that the original B function follows a polynomial manner!

With this new information, we can change B's formula and attempt one with a polynomial model. By tweaking the generalized formula, we end up with the following functions:
- B(A) = 1.5A<sup>2</sup>
- C(A) = 1.55<sup>A</sup>

```python
#Changing the plots to log-log gives us interesting results.
#The B function changed into a linear version! Therefore it is polynomial!
#The C function on the other hand, remained exponential, meaning that it is truly exponential.
plt.figure(figsize=(20, 6))
plt.subplot(231)
plt.title("Log-Log Function B(A)")
plt.scatter(A, B)
plt.plot(A, [1.5*i**2 for i in A], color="c")
plt.xscale("log")
plt.yscale("log")

plt.subplot(132)
plt.title("Log-Log Function C(A)")
plt.scatter(A, C, color="r")
plt.plot(A, [1.55**i for i in A], color="c")
plt.xscale("log")
plt.yscale("log")

plt.subplot(133)
plt.title("Log-Log Both Functions")
plt.scatter(A, B)
plt.scatter(A, C, color="r")
plt.plot(A, [1.55**i for i in A], color="c")
plt.plot(A, [1.5*i**2 for i in A], color="c")
plt.xscale("log")
plt.yscale("log")
```

## Conclusion
After some further tweaking on the normal plottted axis, we end up with B being polynimial and C exponential.
We also end up with the following formulas:

- B(A) = 1.55A<sup>2</sup>
- C(A) = 1.55<sup>A</sup>

**Note: The formulas were not part of the exercise, merely a fun approximation**
